## Task G: Contrast sensitivity function of V1 cell

### 1.1 Define $K(x,y)$ and $K'(x,y)$

$$K(x, y) = \exp{\left( -\frac{x^2}{2\sigma_x^2} - \frac{y^2}{2\sigma_y^2} \right)} \cos(\hat{k}x + \phi)$$



```python
import os
import matplotlib.pyplot as plt
import numpy as np
from math import exp, cos, sin, pi

figures_dir = image_path = os.path.join("..", "figures")

def K(x,y,sigma_x,sigma_y,k_hat,phi):
    '''Gaussian kernel function (V1)'''
    # x = x/2
    # y = y/2
    # return cos(k_hat*x + phi)
    return exp(-(x**2/(2*sigma_x**2) + y**2/(2*sigma_y**2))) * cos(k_hat*x + phi)

def get_kernel_DoG(sigma_x,sigma_y,k_hat,phi,L=None):
    '''Difference of Gaussian kernel (without extensions)'''
    if L is None:
        L = 6 * max(sigma_x, sigma_y)   # size of the kernel, cover [-3sigma, 3sigma]
        print(f"Kernel size L is set to {L}")
    M = np.zeros((L, L))
    for i in range(L):
        y = L/2 - i
        for j in range(L):
            x = j - L/2
            M[i,j] = K(x,y,sigma_x,sigma_y,k_hat,phi)
    return M
```

```python
def g_c_DoG(k, kernel):
    g = 0
    L = kernel.shape[0]
    for i in range(L):
        for j in range(kernel.shape[1]):
            x = j - L/2
            g += kernel[i,j] * cos(k * x)
    return g
def g_s_DoG(k, kernel):
    g = 0
    L = kernel.shape[0]
    for i in range(L):
        for j in range(kernel.shape[1]):
            x = j - L/2
            g += kernel[i,j] * sin(k * x)
    return g

def g_DoG(k, kernel):
    gc = g_c_DoG(k, kernel)
    gs = g_s_DoG(k, kernel)
    return np.sqrt(gc**2 + gs**2)
```

### 1.2 Plot the receptive field

```python
sigma_x, sigma_y, k_hat, phi =2, 3, 2/3*pi, 0
L = np.int32(6 * max(sigma_x, sigma_y))   # size of the kernel, cover [-3sigma, 3sigma]
print(f"Kernel size L is set to {L}")

# get the DoG kernel
kernel = get_kernel_DoG(sigma_x,sigma_y,k_hat,phi,L)
plt.figure(figsize=(10, 4))
plt.imshow(kernel, cmap='bwr', vmax=1, vmin=-1)
plt.title(f"V1 kernel\n($\\sigma_x$= {sigma_x:.3f}, $\\sigma_y$={sigma_y:.3f}, $\\hat k$= {k_hat:.3f}, $\\phi$={phi:.3f})")
plt.colorbar(fraction=0.046, pad=0.04)
plt.tight_layout()
plt.axis('off')
plt.show()
```

### 1.3 plot the contrast sensitivity function

```python
sigma_x,sigma_y,k_hat,phi = 2, 3, 2*pi/3, 0
L = np.int32(6 * max(sigma_x, sigma_y))   # size of the kernel, cover [-3sigma, 3sigma]
kernel = get_kernel_DoG(sigma_x,sigma_y,k_hat,phi,L)


k_list = [n*2*pi /(2*L) for n in range(L+1)]

import bisect
bisect.insort(k_list, k_hat)    # insert the k_hat

g_list = [g_DoG(k, kernel) for k in k_list]


# plot the contrast sensitivity function
plt.figure(figsize=(6,4))
plt.plot(k_list, g_list, marker='.')
plt.xlabel("Spatial frequency k (rad/unit distance)")
plt.ylabel("Response g(k)")
plt.title("Contrast Sensitivity Function of V1 Receptive Field")
plt.grid()
# ---  ---
max_idx = np.argmax(g_list)
max_k = k_list[max_idx]
max_g = g_list[max_idx]
plt.scatter(max_k, max_g, color='red', zorder=2)
plt.text(max_k-0.2, max_g-0.3, f"max={max_g:.2f}\n@k={max_k:.2f}", 
         fontsize=9, color='red', ha='left', va='top')

# 
plt.axvline(x=k_hat, color='green', linestyle='--', linewidth=1)
plt.text(k_hat+0.05, 1, fr"$\hat{{k}} = {k_hat:.2f}$", 
         fontsize=9, color='green', ha='left', va='top')


plt.savefig(os.path.join(figures_dir, f"V1_CSF_sx{sigma_x}_sy{sigma_y}_khat{k_hat:.3f}_phi{phi:.3f}.pdf"), bbox_inches='tight', pad_inches=0.1)
plt.show()
```

```python
for sigma_x in np.arange(0.5,3.5,0.5):
    kernel = get_kernel_DoG(sigma_x, sigma_y, k_hat, phi, L)
    g_list = [g_DoG(k, kernel) for k in k_list]
    plt.plot(k_list, g_list, label=f'sigma_x={sigma_x}')
plt.axvline(k_hat, color='k', linestyle='--', label='k_hat')
plt.title("CSF with different $\\sigma_x$")
plt.xlabel("$k$")
plt.ylabel("$g(k)$")
plt.legend()
plt.savefig(os.path.join(figures_dir, f"V1_CSF_different_sigma_x.pdf"), bbox_inches='tight', pad_inches=0.1)
plt.show()
```
